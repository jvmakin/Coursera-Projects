```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Load Data

```python
sales = pd.read_csv('home_data.csv')
```

# Plotting Square Feet v. Price

```python
sales.plot(x="sqft_living", y="price", kind = 'scatter')
```

# Create One Variable Regression Model

```python
# 1: Split into testing/Training
from sklearn.model_selection import train_test_split

train, test = train_test_split(sales, test_size=0.3)
```

```python
from sklearn import linear_model, datasets
from sklearn.metrics import mean_squared_error, r2_score
import math
# 2. Build Regression Model w/ Square Feet
# Set up X and Y
X_train = train['sqft_living'].values.reshape((-1,1))
Y_train = train['price'].values.reshape((-1,1))

X_test = test['sqft_living'].values.reshape((-1,1))
Y_test = test['price'].values.reshape((-1,1))

# Create linear regression object
regr = linear_model.LinearRegression()

# Train regression object
regr.fit(X_train, Y_train)

# Create Predictions
Y_pred = regr.predict(X_test)

# Print Coefficients etc.
print regr.coef_
print math.pow(mean_squared_error(Y_test, Y_pred), .5)

plt.plot(sales['sqft_living'], sales['price'],'.',
       X_test,Y_pred,'-')
```
